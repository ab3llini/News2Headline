```python
import pandas as pd 
import pickle
```

```python
df = pd.read_pickle('bloomberg_tweets_2018-10-03.pkl')
df_techcrunch = pd.read_pickle('techcrunch_tweets_2018-10-03.pkl')
```

```python
data = list(df_techcrunch[2])
```

```python
import re
link_list = []
for element in data:  
    urls = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-&(-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', str(element))
    link_list.append(urls)
print(link_list)
```

```python
import requests
web_link = 'https://techcrunch.com'
for element in link_list:
    for link in element:
        final_link = requests.get(link).url
        if web_link in final_link:
            print(final_link)
```
